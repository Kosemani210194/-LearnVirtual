```python
#A bar chart showing the number of COVID-19 reported cases in each Nigerian states


plt.style.use('seaborn-pastel')

x = Data.States
y = Data.Laboratory_Confirmed_Cases / 10 ** 2

plt.figure(figsize=(15,30))
plt.barh(x,y, color='red', align='center', alpha=0.65, edgecolor='black')
plt.xlabel('States', fontsize=15)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.ylabel('Laboratory Confirmed Cases', fontsize=15)
plt.title('A Bar Chart Showing the Number of COVID-19 Reported Cases in each Nigerian States', fontsize=17)
plt.savefig('COVID-19 Report in Nigeria', dpi=300)
```

```python
#Exploriing How To Create a Pie Chart

States = ['Lagos', 'Oyo', 'Ogun', 'Osun', 'Ondo']
Laboratory_Confirmed_Cases = [13626, 2219, 1184, 359, 964]
plt.figure(figsize=(8,15))
exp = [0.05,0.04,0.04,0.04,0.04]
cl = ['cyan', 'green', 'red', 'blue', 'orange']
plt.title('COVID-19 Cases in the South-Western States of Nigeria')
plt.pie(Laboratory_Confirmed_Cases, labels = States, explode = exp, autopct = '%2.1f%%', colors = cl)

plt.show()
```

### A program on Histogram showing Blood Sugar Level in some selected men and women

```python
#A simple histogram showing the blood sugar level in some selected men and women

data = {'blood_sugar_men' : [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129], 'blood_sugar_women' : [67, 98, 89, 120, 133, 150, 84, 69, 89, 79,120, 112, 100]}
Data = pd.DataFrame(data)
Data
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]


plt.style.use('seaborn')
plt.figure(figsize=(10,10))
plt.xlabel('Sugar Range in mmol/L', fontsize=10)
plt.ylabel('Total Number of Patients', fontsize=10)
plt.title('Blood Sugar Analysis', fontsize=15)
bins = [60,70,80,90,100,110,120,130,140,150]
plt.hist([blood_sugar_men, blood_sugar_women], bins=bins, rwidth=0.95, color=['green', 'orange'], edgecolor='black')
plt.legend(['Men Blood Sugar', 'Women Blood Sugar'])
plt.show()
```
